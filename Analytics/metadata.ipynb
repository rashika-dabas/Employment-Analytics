```python
# Categorize by Salary Range
def categorize_salary(salary_range):
    if pd.isna(salary_range):
        return 'Unknown'
    salary_values = re.findall(r'\d+', salary_range.replace('$', '').replace(',', ''))
    if len(salary_values) < 2:
        return 'Unknown'
    low, high = int(salary_values[0]), int(salary_values[1])
    avg = (low + high) / 2
    if avg < 50000:
        return 'Entry-level'
    elif 50000 <= avg < 80000:
        return 'Mid-level'
    else:
        return 'Senior-level'

df['salary_category'] = df['career_salary_range'].apply(categorize_salary)

# Recognize Salary Range
def recognize_salary(salary_range):
    if pd.isna(salary_range):
        return (0, 0, 0)
    salary_values = re.findall(r'\d+', salary_range.replace('$', '').replace(',', ''))
    if len(salary_values) < 2:
        return (0, 0, 0)
    low, high = int(salary_values[0]), int(salary_values[1])
    avg = (low + high) / 2
    return (low, high, avg)

df[['min_salary', 'max_salary', 'average_salary']] = df['career_salary_range'].apply(
    lambda x: pd.Series(recognize_salary(x))
)

# Categorize by Industry Sector (keyword matching)
def categorize_industry(job_duties):
    if 'health' in job_duties.lower() or 'medical' in job_duties.lower():
        return 'Healthcare'
    elif 'tech' in job_duties.lower() or 'software' in job_duties.lower():
        return 'Technology'
    elif 'teach' in job_duties.lower() or 'education' in job_duties.lower():
        return 'Education'
    elif 'finance' in job_duties.lower() or 'account' in job_duties.lower():
        return 'Finance'
    else:
        return 'Other'

df['industry_category'] = df['job_duties'].apply(categorize_industry)

df.head()
```

```python
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")
```

```python
# Define keywords for transferable and non-transferable skills
transferable_keywords = [
    'communication', 'teamwork', 'problem-solving', 'critical thinking', 'time management',
    'leadership', 'adaptability', 'interpersonal skills', 'organizational skills', 'analytical skills',
    'project management', 'customer service', 'negotiation', 'writing', 'presentation', 'collaboration',
    'research', 'multitasking'
]

non_transferable_keywords = [
    'coding', 'modeling', 'legal support', 'legal documents', 'equipment', 'machines'
    'cpa', 'machining', 'laboratory', 'radiology', 'document'
    'creativity', 'design', 'advertising', 'marketing', 'chromatography', 'traffic control', 'financial'
    'spectroscopy', 'welding', 'carpentry', 'gdpr', 'methodology', 'control'
]

# Function to extract skills from a paragraph
def extract_skills(paragraph, keywords):
    # Process the paragraph with spaCy
    doc = nlp(paragraph)
    
    # Extract noun phrases that might be skills
    skills = [chunk.text.lower() for chunk in doc.noun_chunks if any(keyword in chunk.text.lower() for keyword in keywords)]
    
    return skills

# Apply the function to the 'Career Description' column
df['transferable_skills'] = df['career_description'].apply(lambda x: extract_skills(x, transferable_keywords))
df['non_transferable_skills'] = df['career_description'].apply(lambda x: extract_skills(x, non_transferable_keywords))

# Display the results
for index, row in df.iterrows():
    print(f"Job Title: {row['career_name']}")
    print(f"Transferable Skills: {row['transferable_skills']}")
    print(f"Non-Transferable Skills: {row['non_transferable_skills']}")
    print("---")
```

```python
# Extract the 'Grade 11 and 12 Subjects' column
subjects_series = df['grade_11_12_subjects']

# Function to process and extract unique subjects
def extract_subjects(subjects_series):
    subjects_list = []
    for subjects in subjects_series:
        subjects_list.extend(subjects.split(', '))
    subjs = [subject.split(' and ') for subject in subjects_list]
    subjects = [s.replace('and ', '') for sub in subjs for s in sub]
    subjects = [sub.replace(' and', '') for sub in subjects]
    return list(set(subjects))

# Extract unique subjects
unique_subjects = extract_subjects(subjects_series)

# Display the unique subjects
print("Unique Grade 11 12 Subjects:")
for subject in unique_subjects:
    print(subject)

# Analyze the correlation between subjects and careers
subject_career_mapping = {}
for index, row in df.iterrows():
    if pd.notna(row['grade_11_12_subjects']):
        subjects_list = row['grade_11_12_subjects'].split(', ')
        subjs = [subject.split(' and ') for subject in subjects_list]
        subjects = [s.replace('and ', '') for sub in subjs for s in sub]
        subjects = [sub.replace(' and', '') for sub in subjects]
        for subject in subjects:
            if subject not in subject_career_mapping:
                subject_career_mapping[subject] = []
            subject_career_mapping[subject].append(row['career_name'])

# Display the subject-career mapping
print("\nSubject-Career Mapping:")
for subject, careers in subject_career_mapping.items():
    print(f"{subject}: {', '.join(careers)}")
```

```python
# Store the subject-career mapping in a DataFrame
subject_career_df = pd.DataFrame(subject_career_mapping.items(), columns=['Subject', 'Careers'])
subject_career_df['Number of Careers'] = subject_career_df['Careers'].apply(len)
subject_career_df
```

```python
# Function to process and extract unique subjects
def extract_subjects(subjects):
    subjects_list = []
    subjects = subjects.replace("Bachelor's Degree in ", '')
    subjects = subjects.replace("Master's Degree in ", '')
    subjects = subjects.replace(" or ", ', ')
    subjects = subjects.replace(",, ", ', ')
    subjects = subjects.replace(" (Minimum)", ', ')
    subjects = subjects.replace(" (Nice to have)", ', ')
    subjects_list.extend(subjects.split(', '))
    if '' in subjects_list: subjects_list.remove('')
    subjects_list = list(set(subjects_list))
    return subjects_list

# Join value from two columns
df['university_subjects'] = df['post_secondary_degree'] + df['advanced_post_secondary_degree']

df['university_subjects'] = df['university_subjects'].apply(extract_subjects)

df['university_subjects']
```

```python
def extract_relevant_info(text):
    if pd.isna(text):
        return "No information available"
    
    # Convert to lowercase for easier processing
    text = text.lower()
    
    # Define keywords to look for
    keywords = ['growth', 'demand', 'increase', 'decrease', 'trend', 'projection']
    
    # Extract sentences containing keywords
    relevant_sentences = []
    sentences = text.split('.')
    for sentence in sentences:
        if any(keyword in sentence for keyword in keywords):
            relevant_sentences.append(sentence.strip().capitalize())
    
    return '. '.join(relevant_sentences)+'.' if relevant_sentences else "No relevant information found."

# Apply the function to the Future column
df['extracted_future_info'] = df['future'].apply(extract_relevant_info)

# Display the results
for index, row in df.iterrows():
    print(f"Job Title: {row['career_name']}")
    print(f"Extracted Future Info: {row['extracted_future_info']}")
    print("---")
```
